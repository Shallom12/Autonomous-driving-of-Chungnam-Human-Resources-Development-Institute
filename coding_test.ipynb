#1. 👁️ 인지 (Perception): 객체 감지 시뮬레이션
##차량의 카메라가 주변 객체(예: 다른 차량, 보행자)를 감지하는 과정을 단순화한 코드입니다. 감지된 객체의 종류와 위치를 출력합니다.

```python
import random

def simulate_object_detection():
    """
    주변 객체 감지를 시뮬레이션합니다.
    (실제로는 카메라/센서 데이터와 딥러닝 모델 사용)
    """
    detected_objects = []

    # 랜덤하게 객체들을 생성합니다 (실제로는 센서가 감지)
    num_objects = random.randint(1, 5) # 1개에서 5개 사이의 객체 감지

    object_types = ["차량", "보행자", "신호등", "자전거", "오토바이"]

    print("--- 인지 모듈: 객체 감지 시작 ---")
    for i in range(num_objects):
        obj_type = random.choice(object_types)
        distance = round(random.uniform(2, 50), 1) # 2m ~ 50m 거리
        angle = round(random.uniform(-45, 45), 1) # -45도 ~ +45도 시야각

        detected_objects.append({
            "type": obj_type,
            "distance": distance,
            "angle": angle # 차량 진행 방향 기준
        })
        print(f"- {obj_type} 감지: {distance:.1f}m 전방, {angle:.1f}도 방향")

    print(f"총 {len(detected_objects)}개의 객체가 감지되었습니다.")
    print("--- 인지 모듈: 객체 감지 완료 ---")
    return detected_objects

# 인지 모듈 실행
detected_items = simulate_object_detection()
# print(detected_items) # 감지된 객체 목록 확인 (선택 사항)
```

# 2.📍 측위 (Localization): 단순화된 차량 위치 추정
## 차량이 지도 위에서 자신의 위치를 추정하는 과정을 단순화했습니다. GPS와 같은 센서 데이터의 오차를 보정하여 더 정확한 위치를 얻는 개념을 보여줍니다.

```python
def simulate_localization(initial_gps_lat, initial_gps_lon):
    """
    차량의 현재 위치를 추정하는 과정을 시뮬레이션합니다.
    (실제로는 GPS, IMU, HD Map 매칭 등 복합 기술 사용)
    """
    print("\n--- 측위 모듈: 차량 위치 추정 시작 ---")

    # 1. GPS를 통한 초기 위치 획득 (약간의 오차 포함)
    gps_lat = initial_gps_lat + random.uniform(-0.0001, 0.0001) # 위도 오차 추가
    gps_lon = initial_gps_lon + random.uniform(-0.0001, 0.0001) # 경도 오차 추가
    print(f"1. GPS 초기 위치: 위도 {gps_lat:.6f}, 경도 {gps_lon:.6f}")

    # 2. 고정밀 지도(HD Map) 또는 다른 센서로 위치 보정
    # 실제로는 주변 특징점(건물, 가로등)과 HD Map을 매칭하여 보정
    # 여기서는 간단히 오차를 줄이는 것으로 시뮬레이션

    corrected_lat = (gps_lat + initial_gps_lat) / 2 # 실제 위치에 더 가깝게 보정
    corrected_lon = (gps_lon + initial_gps_lon) / 2 # 실제 위치에 더 가깝게 보정

    print(f"2. 센서 퓨전/HD Map 보정 후 위치: 위도 {corrected_lat:.6f}, 경도 {corrected_lon:.6f}")

    print("--- 측위 모듈: 차량 위치 추정 완료 ---")
    return corrected_lat, corrected_lon

# 서울 시청 근처의 가상 위경도 (실제 위치)
true_lat, true_lon = 37.5665, 126.9780

# 측위 모듈 실행
current_vehicle_position = simulate_localization(true_lat, true_lon)
# print(current_vehicle_position)
```

#3. 🧠 판단/계획 (Decision Making / Planning): 신호등 및 장애물 반응 시뮬레이션
##차량이 인지 및 측위 정보를 바탕으로 어떤 행동을 할지 판단하고, 그 행동을 어떻게 실행할지 계획하는 과정을 보여줍니다

```python
def make_decision_and_plan(detected_objects, current_speed_kmh, traffic_light_color="green"):
    """
    인지된 객체와 현재 상황을 기반으로 주행 행동을 판단하고 경로를 계획합니다.
    """
    print("\n--- 판단/계획 모듈: 행동 결정 및 경로 계획 시작 ---")

    # 1. 주변 상황 분석 및 판단
    decision = "주행 계속"
    target_speed_kmh = current_speed_kmh

    # 신호등 판단
    if traffic_light_color == "red":
        decision = "정지"
        target_speed_kmh = 0
        print(f" - 신호등이 빨간색입니다. {decision} 결정.")
    elif traffic_light_color == "yellow":
        if current_speed_kmh > 10: # 노란 불에 급정거 위험 시 통과 시도 (고도화된 로직 필요)
            decision = "주의하며 통과"
            target_speed_kmh = 30 # 속도 조절
            print(f" - 신호등이 노란색입니다. {decision} 결정 (현재 속도: {current_speed_kmh} km/h).")
        else:
            decision = "정지"
            target_speed_kmh = 0
            print(f" - 신호등이 노란색입니다. {decision} 결정.")
    else: # green
        print(f" - 신호등이 초록색입니다. 주행 계속.")

    # 장애물 판단
    for obj in detected_objects:
        if obj["type"] in ["차량", "보행자", "자전거"] and obj["distance"] < 15:
            if obj["distance"] < 5: # 매우 가까이 있는 경우
                decision = "긴급 정지"
                target_speed_kmh = 0
                print(f" - !!! 경고: {obj['distance']:.1f}m 전방에 {obj['type']} 감지. {decision} 결정.")
                break # 가장 가까운 위험 객체에만 반응하도록
            elif decision != "정지": # 이미 정지 결정이 아니면 속도 줄임
                decision = "속도 감속 및 안전거리 확보"
                target_speed_kmh = min(target_speed_kmh, current_speed_kmh * 0.5) # 현재 속도의 절반으로 줄임
                print(f" - {obj['distance']:.1f}m 전방에 {obj['type']} 감지. {decision} 결정.")

    # 2. 경로 계획 (결정에 따른 단순화된 계획)
    if decision == "정지" or decision == "긴급 정지":
        planned_action = "브레이크 작동, 차량 정지"
        planned_trajectory = "현재 위치에서 정지"
    elif decision == "속도 감속 및 안전거리 확보":
        planned_action = f"속도 {target_speed_kmh:.0f}km/h로 감속, 전방 주시"
        planned_trajectory = "전방 객체와의 안전거리 확보하며 주행"
    elif decision == "주의하며 통과":
        planned_action = f"속도 {target_speed_kmh:.0f}km/h로 조절, 교차로 통과"
        planned_trajectory = "교차로 안전하게 통과"
    else: # 주행 계속
        planned_action = "현재 속도 유지 또는 가속"
        planned_trajectory = "차선 유지하며 직진"

    print(f"\n최종 결정: {decision}")
    print(f"계획된 행동: {planned_action}")
    print(f"계획된 궤적: {planned_trajectory}")

    print("--- 판단/계획 모듈: 행동 결정 및 경로 계획 완료 ---")
    return decision, planned_action, target_speed_kmh

# 판단/계획 모듈 실행 예시
# 예시 1: 전방에 차량이 있고 신호등은 초록색
# detected_obj_1 = [{"type": "차량", "distance": 10.0, "angle": 0}]
# current_speed_1 = 60
# decision_1, action_1, speed_1 = make_decision_and_plan(detected_obj_1, current_speed_1, "green")

# 예시 2: 신호등이 빨간색이고 전방에 아무것도 없음
# detected_obj_2 = []
# current_speed_2 = 40
# decision_2, action_2, speed_2 = make_decision_and_plan(detected_obj_2, current_speed_2, "red")

# 예시 3: 전방에 보행자가 매우 가깝게 나타남
detected_obj_3 = [{"type": "보행자", "distance": 3.0, "angle": 5}]
current_speed_3 = 30
decision_3, action_3, speed_3 = make_decision_and_plan(detected_obj_3, current_speed_3, "green")
```

#4. 🕹️ 제어 (Control): 속도 및 조향 제어 시뮬레이션
##판단/계획 모듈에서 결정된 목표 속도와 조향 목표를 바탕으로, 차량의 실제 속도와 조향 각도를 조절하는 과정을 시뮬레이션합니다. 여기서는 목표값에 도달하기 위한 간단한 제어 루프를 보여줍니다.



```python
import time

def simulate_control(target_speed_kmh, target_steering_angle_deg, current_speed_kmh=0, current_steering_angle_deg=0):
    """
    차량의 속도와 조향을 제어하여 목표 값에 도달하도록 시뮬레이션합니다.
    (실제로는 PID 제어, MPC 등 정교한 제어 알고리즘 사용)
    """
    print("\n--- 제어 모듈: 차량 제어 시작 ---")
    print(f"목표 속도: {target_speed_kmh:.0f} km/h, 목표 조향 각도: {target_steering_angle_deg:.1f} 도")

    # 속도 제어
    speed_error = target_speed_kmh - current_speed_kmh
    print(f"현재 속도: {current_speed_kmh:.1f} km/h, 속도 오차: {speed_error:.1f} km/h")

    # 조향 제어
    steering_error = target_steering_angle_deg - current_steering_angle_deg
    print(f"현재 조향: {current_steering_angle_deg:.1f} 도, 조향 오차: {steering_error:.1f} 도")

    steps = 0
    max_steps = 5 # 5단계에 걸쳐 목표에 도달한다고 가정

    while (abs(speed_error) > 0.5 or abs(steering_error) > 0.5) and steps < max_steps:
        steps += 1
        print(f"\n--- 제어 루프 {steps} ---")

        # --- 속도 제어 ---
        # 간단한 비례 제어 (오차의 일정 비율만큼 속도 조절)
        speed_control_output = speed_error * 0.3 # 제어 상수
        current_speed_kmh += speed_control_output

        # 속도 오차 업데이트
        speed_error = target_speed_kmh - current_speed_kmh
        print(f"-> 속도 제어: 현재 속도 {current_speed_kmh:.1f} km/h (오차: {speed_error:.1f} km/h)")

        # --- 조향 제어 ---
        # 간단한 비례 제어 (오차의 일정 비율만큼 조향 조절)
        steering_control_output = steering_error * 0.2 # 제어 상수
        current_steering_angle_deg += steering_control_output

        # 조향 오차 업데이트
        steering_error = target_steering_angle_deg - current_steering_angle_deg
        print(f"-> 조향 제어: 현재 조향 {current_steering_angle_deg:.1f} 도 (오차: {steering_error:.1f} 도)")

        time.sleep(0.1) # 시뮬레이션 지연 (실시간 제어 시뮬레이션)

    print("\n--- 제어 모듈: 목표에 도달 (또는 근접) 완료 ---")
    print(f"최종 속도: {current_speed_kmh:.1f} km/h, 최종 조향 각도: {current_steering_angle_deg:.1f} 도")
    return current_speed_kmh, current_steering_angle_deg

# 제어 모듈 실행 예시
# 예시 1: 50km/h로 가속하고 핸들 5도 돌리기
simulate_control(target_speed_kmh=50, target_steering_angle_deg=5)

# 예시 2: 0km/h로 정지하고 핸들 직진
# simulate_control(target_speed_kmh=0, target_steering_angle_deg=0, current_speed_kmh=30, current_steering_angle_deg=10)
```
