#1. ğŸ‘ï¸ ì¸ì§€ (Perception): ê°ì²´ ê°ì§€ ì‹œë®¬ë ˆì´ì…˜
##ì°¨ëŸ‰ì˜ ì¹´ë©”ë¼ê°€ ì£¼ë³€ ê°ì²´(ì˜ˆ: ë‹¤ë¥¸ ì°¨ëŸ‰, ë³´í–‰ì)ë¥¼ ê°ì§€í•˜ëŠ” ê³¼ì •ì„ ë‹¨ìˆœí™”í•œ ì½”ë“œì…ë‹ˆë‹¤. ê°ì§€ëœ ê°ì²´ì˜ ì¢…ë¥˜ì™€ ìœ„ì¹˜ë¥¼ ì¶œë ¥í•©ë‹ˆë‹¤.

```python
import random

def simulate_object_detection():
    """
    ì£¼ë³€ ê°ì²´ ê°ì§€ë¥¼ ì‹œë®¬ë ˆì´ì…˜í•©ë‹ˆë‹¤.
    (ì‹¤ì œë¡œëŠ” ì¹´ë©”ë¼/ì„¼ì„œ ë°ì´í„°ì™€ ë”¥ëŸ¬ë‹ ëª¨ë¸ ì‚¬ìš©)
    """
    detected_objects = []

    # ëœë¤í•˜ê²Œ ê°ì²´ë“¤ì„ ìƒì„±í•©ë‹ˆë‹¤ (ì‹¤ì œë¡œëŠ” ì„¼ì„œê°€ ê°ì§€)
    num_objects = random.randint(1, 5) # 1ê°œì—ì„œ 5ê°œ ì‚¬ì´ì˜ ê°ì²´ ê°ì§€

    object_types = ["ì°¨ëŸ‰", "ë³´í–‰ì", "ì‹ í˜¸ë“±", "ìì „ê±°", "ì˜¤í† ë°”ì´"]

    print("--- ì¸ì§€ ëª¨ë“ˆ: ê°ì²´ ê°ì§€ ì‹œì‘ ---")
    for i in range(num_objects):
        obj_type = random.choice(object_types)
        distance = round(random.uniform(2, 50), 1) # 2m ~ 50m ê±°ë¦¬
        angle = round(random.uniform(-45, 45), 1) # -45ë„ ~ +45ë„ ì‹œì•¼ê°

        detected_objects.append({
            "type": obj_type,
            "distance": distance,
            "angle": angle # ì°¨ëŸ‰ ì§„í–‰ ë°©í–¥ ê¸°ì¤€
        })
        print(f"- {obj_type} ê°ì§€: {distance:.1f}m ì „ë°©, {angle:.1f}ë„ ë°©í–¥")

    print(f"ì´ {len(detected_objects)}ê°œì˜ ê°ì²´ê°€ ê°ì§€ë˜ì—ˆìŠµë‹ˆë‹¤.")
    print("--- ì¸ì§€ ëª¨ë“ˆ: ê°ì²´ ê°ì§€ ì™„ë£Œ ---")
    return detected_objects

# ì¸ì§€ ëª¨ë“ˆ ì‹¤í–‰
detected_items = simulate_object_detection()
# print(detected_items) # ê°ì§€ëœ ê°ì²´ ëª©ë¡ í™•ì¸ (ì„ íƒ ì‚¬í•­)
```

# 2.ğŸ“ ì¸¡ìœ„ (Localization): ë‹¨ìˆœí™”ëœ ì°¨ëŸ‰ ìœ„ì¹˜ ì¶”ì •
## ì°¨ëŸ‰ì´ ì§€ë„ ìœ„ì—ì„œ ìì‹ ì˜ ìœ„ì¹˜ë¥¼ ì¶”ì •í•˜ëŠ” ê³¼ì •ì„ ë‹¨ìˆœí™”í–ˆìŠµë‹ˆë‹¤. GPSì™€ ê°™ì€ ì„¼ì„œ ë°ì´í„°ì˜ ì˜¤ì°¨ë¥¼ ë³´ì •í•˜ì—¬ ë” ì •í™•í•œ ìœ„ì¹˜ë¥¼ ì–»ëŠ” ê°œë…ì„ ë³´ì—¬ì¤ë‹ˆë‹¤.

```python
def simulate_localization(initial_gps_lat, initial_gps_lon):
    """
    ì°¨ëŸ‰ì˜ í˜„ì¬ ìœ„ì¹˜ë¥¼ ì¶”ì •í•˜ëŠ” ê³¼ì •ì„ ì‹œë®¬ë ˆì´ì…˜í•©ë‹ˆë‹¤.
    (ì‹¤ì œë¡œëŠ” GPS, IMU, HD Map ë§¤ì¹­ ë“± ë³µí•© ê¸°ìˆ  ì‚¬ìš©)
    """
    print("\n--- ì¸¡ìœ„ ëª¨ë“ˆ: ì°¨ëŸ‰ ìœ„ì¹˜ ì¶”ì • ì‹œì‘ ---")

    # 1. GPSë¥¼ í†µí•œ ì´ˆê¸° ìœ„ì¹˜ íšë“ (ì•½ê°„ì˜ ì˜¤ì°¨ í¬í•¨)
    gps_lat = initial_gps_lat + random.uniform(-0.0001, 0.0001) # ìœ„ë„ ì˜¤ì°¨ ì¶”ê°€
    gps_lon = initial_gps_lon + random.uniform(-0.0001, 0.0001) # ê²½ë„ ì˜¤ì°¨ ì¶”ê°€
    print(f"1. GPS ì´ˆê¸° ìœ„ì¹˜: ìœ„ë„ {gps_lat:.6f}, ê²½ë„ {gps_lon:.6f}")

    # 2. ê³ ì •ë°€ ì§€ë„(HD Map) ë˜ëŠ” ë‹¤ë¥¸ ì„¼ì„œë¡œ ìœ„ì¹˜ ë³´ì •
    # ì‹¤ì œë¡œëŠ” ì£¼ë³€ íŠ¹ì§•ì (ê±´ë¬¼, ê°€ë¡œë“±)ê³¼ HD Mapì„ ë§¤ì¹­í•˜ì—¬ ë³´ì •
    # ì—¬ê¸°ì„œëŠ” ê°„ë‹¨íˆ ì˜¤ì°¨ë¥¼ ì¤„ì´ëŠ” ê²ƒìœ¼ë¡œ ì‹œë®¬ë ˆì´ì…˜

    corrected_lat = (gps_lat + initial_gps_lat) / 2 # ì‹¤ì œ ìœ„ì¹˜ì— ë” ê°€ê¹ê²Œ ë³´ì •
    corrected_lon = (gps_lon + initial_gps_lon) / 2 # ì‹¤ì œ ìœ„ì¹˜ì— ë” ê°€ê¹ê²Œ ë³´ì •

    print(f"2. ì„¼ì„œ í“¨ì „/HD Map ë³´ì • í›„ ìœ„ì¹˜: ìœ„ë„ {corrected_lat:.6f}, ê²½ë„ {corrected_lon:.6f}")

    print("--- ì¸¡ìœ„ ëª¨ë“ˆ: ì°¨ëŸ‰ ìœ„ì¹˜ ì¶”ì • ì™„ë£Œ ---")
    return corrected_lat, corrected_lon

# ì„œìš¸ ì‹œì²­ ê·¼ì²˜ì˜ ê°€ìƒ ìœ„ê²½ë„ (ì‹¤ì œ ìœ„ì¹˜)
true_lat, true_lon = 37.5665, 126.9780

# ì¸¡ìœ„ ëª¨ë“ˆ ì‹¤í–‰
current_vehicle_position = simulate_localization(true_lat, true_lon)
# print(current_vehicle_position)
```

#3. ğŸ§  íŒë‹¨/ê³„íš (Decision Making / Planning): ì‹ í˜¸ë“± ë° ì¥ì• ë¬¼ ë°˜ì‘ ì‹œë®¬ë ˆì´ì…˜
##ì°¨ëŸ‰ì´ ì¸ì§€ ë° ì¸¡ìœ„ ì •ë³´ë¥¼ ë°”íƒ•ìœ¼ë¡œ ì–´ë–¤ í–‰ë™ì„ í• ì§€ íŒë‹¨í•˜ê³ , ê·¸ í–‰ë™ì„ ì–´ë–»ê²Œ ì‹¤í–‰í• ì§€ ê³„íší•˜ëŠ” ê³¼ì •ì„ ë³´ì—¬ì¤ë‹ˆë‹¤

```python
def make_decision_and_plan(detected_objects, current_speed_kmh, traffic_light_color="green"):
    """
    ì¸ì§€ëœ ê°ì²´ì™€ í˜„ì¬ ìƒí™©ì„ ê¸°ë°˜ìœ¼ë¡œ ì£¼í–‰ í–‰ë™ì„ íŒë‹¨í•˜ê³  ê²½ë¡œë¥¼ ê³„íší•©ë‹ˆë‹¤.
    """
    print("\n--- íŒë‹¨/ê³„íš ëª¨ë“ˆ: í–‰ë™ ê²°ì • ë° ê²½ë¡œ ê³„íš ì‹œì‘ ---")

    # 1. ì£¼ë³€ ìƒí™© ë¶„ì„ ë° íŒë‹¨
    decision = "ì£¼í–‰ ê³„ì†"
    target_speed_kmh = current_speed_kmh

    # ì‹ í˜¸ë“± íŒë‹¨
    if traffic_light_color == "red":
        decision = "ì •ì§€"
        target_speed_kmh = 0
        print(f" - ì‹ í˜¸ë“±ì´ ë¹¨ê°„ìƒ‰ì…ë‹ˆë‹¤. {decision} ê²°ì •.")
    elif traffic_light_color == "yellow":
        if current_speed_kmh > 10: # ë…¸ë€ ë¶ˆì— ê¸‰ì •ê±° ìœ„í—˜ ì‹œ í†µê³¼ ì‹œë„ (ê³ ë„í™”ëœ ë¡œì§ í•„ìš”)
            decision = "ì£¼ì˜í•˜ë©° í†µê³¼"
            target_speed_kmh = 30 # ì†ë„ ì¡°ì ˆ
            print(f" - ì‹ í˜¸ë“±ì´ ë…¸ë€ìƒ‰ì…ë‹ˆë‹¤. {decision} ê²°ì • (í˜„ì¬ ì†ë„: {current_speed_kmh} km/h).")
        else:
            decision = "ì •ì§€"
            target_speed_kmh = 0
            print(f" - ì‹ í˜¸ë“±ì´ ë…¸ë€ìƒ‰ì…ë‹ˆë‹¤. {decision} ê²°ì •.")
    else: # green
        print(f" - ì‹ í˜¸ë“±ì´ ì´ˆë¡ìƒ‰ì…ë‹ˆë‹¤. ì£¼í–‰ ê³„ì†.")

    # ì¥ì• ë¬¼ íŒë‹¨
    for obj in detected_objects:
        if obj["type"] in ["ì°¨ëŸ‰", "ë³´í–‰ì", "ìì „ê±°"] and obj["distance"] < 15:
            if obj["distance"] < 5: # ë§¤ìš° ê°€ê¹Œì´ ìˆëŠ” ê²½ìš°
                decision = "ê¸´ê¸‰ ì •ì§€"
                target_speed_kmh = 0
                print(f" - !!! ê²½ê³ : {obj['distance']:.1f}m ì „ë°©ì— {obj['type']} ê°ì§€. {decision} ê²°ì •.")
                break # ê°€ì¥ ê°€ê¹Œìš´ ìœ„í—˜ ê°ì²´ì—ë§Œ ë°˜ì‘í•˜ë„ë¡
            elif decision != "ì •ì§€": # ì´ë¯¸ ì •ì§€ ê²°ì •ì´ ì•„ë‹ˆë©´ ì†ë„ ì¤„ì„
                decision = "ì†ë„ ê°ì† ë° ì•ˆì „ê±°ë¦¬ í™•ë³´"
                target_speed_kmh = min(target_speed_kmh, current_speed_kmh * 0.5) # í˜„ì¬ ì†ë„ì˜ ì ˆë°˜ìœ¼ë¡œ ì¤„ì„
                print(f" - {obj['distance']:.1f}m ì „ë°©ì— {obj['type']} ê°ì§€. {decision} ê²°ì •.")

    # 2. ê²½ë¡œ ê³„íš (ê²°ì •ì— ë”°ë¥¸ ë‹¨ìˆœí™”ëœ ê³„íš)
    if decision == "ì •ì§€" or decision == "ê¸´ê¸‰ ì •ì§€":
        planned_action = "ë¸Œë ˆì´í¬ ì‘ë™, ì°¨ëŸ‰ ì •ì§€"
        planned_trajectory = "í˜„ì¬ ìœ„ì¹˜ì—ì„œ ì •ì§€"
    elif decision == "ì†ë„ ê°ì† ë° ì•ˆì „ê±°ë¦¬ í™•ë³´":
        planned_action = f"ì†ë„ {target_speed_kmh:.0f}km/hë¡œ ê°ì†, ì „ë°© ì£¼ì‹œ"
        planned_trajectory = "ì „ë°© ê°ì²´ì™€ì˜ ì•ˆì „ê±°ë¦¬ í™•ë³´í•˜ë©° ì£¼í–‰"
    elif decision == "ì£¼ì˜í•˜ë©° í†µê³¼":
        planned_action = f"ì†ë„ {target_speed_kmh:.0f}km/hë¡œ ì¡°ì ˆ, êµì°¨ë¡œ í†µê³¼"
        planned_trajectory = "êµì°¨ë¡œ ì•ˆì „í•˜ê²Œ í†µê³¼"
    else: # ì£¼í–‰ ê³„ì†
        planned_action = "í˜„ì¬ ì†ë„ ìœ ì§€ ë˜ëŠ” ê°€ì†"
        planned_trajectory = "ì°¨ì„  ìœ ì§€í•˜ë©° ì§ì§„"

    print(f"\nìµœì¢… ê²°ì •: {decision}")
    print(f"ê³„íšëœ í–‰ë™: {planned_action}")
    print(f"ê³„íšëœ ê¶¤ì : {planned_trajectory}")

    print("--- íŒë‹¨/ê³„íš ëª¨ë“ˆ: í–‰ë™ ê²°ì • ë° ê²½ë¡œ ê³„íš ì™„ë£Œ ---")
    return decision, planned_action, target_speed_kmh

# íŒë‹¨/ê³„íš ëª¨ë“ˆ ì‹¤í–‰ ì˜ˆì‹œ
# ì˜ˆì‹œ 1: ì „ë°©ì— ì°¨ëŸ‰ì´ ìˆê³  ì‹ í˜¸ë“±ì€ ì´ˆë¡ìƒ‰
# detected_obj_1 = [{"type": "ì°¨ëŸ‰", "distance": 10.0, "angle": 0}]
# current_speed_1 = 60
# decision_1, action_1, speed_1 = make_decision_and_plan(detected_obj_1, current_speed_1, "green")

# ì˜ˆì‹œ 2: ì‹ í˜¸ë“±ì´ ë¹¨ê°„ìƒ‰ì´ê³  ì „ë°©ì— ì•„ë¬´ê²ƒë„ ì—†ìŒ
# detected_obj_2 = []
# current_speed_2 = 40
# decision_2, action_2, speed_2 = make_decision_and_plan(detected_obj_2, current_speed_2, "red")

# ì˜ˆì‹œ 3: ì „ë°©ì— ë³´í–‰ìê°€ ë§¤ìš° ê°€ê¹ê²Œ ë‚˜íƒ€ë‚¨
detected_obj_3 = [{"type": "ë³´í–‰ì", "distance": 3.0, "angle": 5}]
current_speed_3 = 30
decision_3, action_3, speed_3 = make_decision_and_plan(detected_obj_3, current_speed_3, "green")
```

```python
import time

def simulate_control(target_speed_kmh, target_steering_angle_deg, current_speed_kmh=0, current_steering_angle_deg=0):
    """
    ì°¨ëŸ‰ì˜ ì†ë„ì™€ ì¡°í–¥ì„ ì œì–´í•˜ì—¬ ëª©í‘œ ê°’ì— ë„ë‹¬í•˜ë„ë¡ ì‹œë®¬ë ˆì´ì…˜í•©ë‹ˆë‹¤.
    (ì‹¤ì œë¡œëŠ” PID ì œì–´, MPC ë“± ì •êµí•œ ì œì–´ ì•Œê³ ë¦¬ì¦˜ ì‚¬ìš©)
    """
    print("\n--- ì œì–´ ëª¨ë“ˆ: ì°¨ëŸ‰ ì œì–´ ì‹œì‘ ---")
    print(f"ëª©í‘œ ì†ë„: {target_speed_kmh:.0f} km/h, ëª©í‘œ ì¡°í–¥ ê°ë„: {target_steering_angle_deg:.1f} ë„")

    # ì†ë„ ì œì–´
    speed_error = target_speed_kmh - current_speed_kmh
    print(f"í˜„ì¬ ì†ë„: {current_speed_kmh:.1f} km/h, ì†ë„ ì˜¤ì°¨: {speed_error:.1f} km/h")

    # ì¡°í–¥ ì œì–´
    steering_error = target_steering_angle_deg - current_steering_angle_deg
    print(f"í˜„ì¬ ì¡°í–¥: {current_steering_angle_deg:.1f} ë„, ì¡°í–¥ ì˜¤ì°¨: {steering_error:.1f} ë„")

    steps = 0
    max_steps = 5 # 5ë‹¨ê³„ì— ê±¸ì³ ëª©í‘œì— ë„ë‹¬í•œë‹¤ê³  ê°€ì •

    while (abs(speed_error) > 0.5 or abs(steering_error) > 0.5) and steps < max_steps:
        steps += 1
        print(f"\n--- ì œì–´ ë£¨í”„ {steps} ---")

        # --- ì†ë„ ì œì–´ ---
        # ê°„ë‹¨í•œ ë¹„ë¡€ ì œì–´ (ì˜¤ì°¨ì˜ ì¼ì • ë¹„ìœ¨ë§Œí¼ ì†ë„ ì¡°ì ˆ)
        speed_control_output = speed_error * 0.3 # ì œì–´ ìƒìˆ˜
        current_speed_kmh += speed_control_output

        # ì†ë„ ì˜¤ì°¨ ì—…ë°ì´íŠ¸
        speed_error = target_speed_kmh - current_speed_kmh
        print(f"-> ì†ë„ ì œì–´: í˜„ì¬ ì†ë„ {current_speed_kmh:.1f} km/h (ì˜¤ì°¨: {speed_error:.1f} km/h)")

        # --- ì¡°í–¥ ì œì–´ ---
        # ê°„ë‹¨í•œ ë¹„ë¡€ ì œì–´ (ì˜¤ì°¨ì˜ ì¼ì • ë¹„ìœ¨ë§Œí¼ ì¡°í–¥ ì¡°ì ˆ)
        steering_control_output = steering_error * 0.2 # ì œì–´ ìƒìˆ˜
        current_steering_angle_deg += steering_control_output

        # ì¡°í–¥ ì˜¤ì°¨ ì—…ë°ì´íŠ¸
        steering_error = target_steering_angle_deg - current_steering_angle_deg
        print(f"-> ì¡°í–¥ ì œì–´: í˜„ì¬ ì¡°í–¥ {current_steering_angle_deg:.1f} ë„ (ì˜¤ì°¨: {steering_error:.1f} ë„)")

        time.sleep(0.1) # ì‹œë®¬ë ˆì´ì…˜ ì§€ì—° (ì‹¤ì‹œê°„ ì œì–´ ì‹œë®¬ë ˆì´ì…˜)

    print("\n--- ì œì–´ ëª¨ë“ˆ: ëª©í‘œì— ë„ë‹¬ (ë˜ëŠ” ê·¼ì ‘) ì™„ë£Œ ---")
    print(f"ìµœì¢… ì†ë„: {current_speed_kmh:.1f} km/h, ìµœì¢… ì¡°í–¥ ê°ë„: {current_steering_angle_deg:.1f} ë„")
    return current_speed_kmh, current_steering_angle_deg

# ì œì–´ ëª¨ë“ˆ ì‹¤í–‰ ì˜ˆì‹œ
# ì˜ˆì‹œ 1: 50km/hë¡œ ê°€ì†í•˜ê³  í•¸ë“¤ 5ë„ ëŒë¦¬ê¸°
simulate_control(target_speed_kmh=50, target_steering_angle_deg=5)

# ì˜ˆì‹œ 2: 0km/hë¡œ ì •ì§€í•˜ê³  í•¸ë“¤ ì§ì§„
# simulate_control(target_speed_kmh=0, target_steering_angle_deg=0, current_speed_kmh=30, current_steering_angle_deg=10)
```
